```python
# Análisando Queatão 7
questao_7 = kaggle_data.loc[:, 'Q7_Part_1':'Q7_OTHER']
questao_7.head()
```

```python
questao_7.columns = list(questao_7.mode().values)
q7 = questao_7.count().reset_index()
q7.columns = ['language', 'count']
q7 = q7.sort_values('count', ascending=False)
q7
```

```python
fig_q7 = px.histogram(q7, x="language", y="count")
fig_q7.show()
```
